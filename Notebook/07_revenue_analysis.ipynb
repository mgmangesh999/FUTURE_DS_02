```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Cleaned-Telco-Customer-Churn.csv")
```

### Total Revenue

```python
total_revenue = df['MonthlyCharges'].sum()
print(total_revenue)
```

### Revenue Lost

```python
churn_revenue = df[df['Churn'] == 'Yes']['MonthlyCharges'].sum()
print(churn_revenue)
```

### Average Customer lifetime


```python
avg_tenure = df['tenure'].mean()
print(avg_tenure)
```
